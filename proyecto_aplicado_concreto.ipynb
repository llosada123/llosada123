```
###--- reemplazo de las letras con tildes por sin tildes.
import csv
import re
from google.colab import files

# Leer el archivo CSV con los datos extraídos
with open('datos_extraidos_total.csv', 'r', encoding='utf-8-sig') as archivo:
    lector_csv = csv.reader(archivo)
    # Extraer los datos de la primera columna en una lista
    datos_extraidos_total = [fila[0] for fila in lector_csv]

# Definir un diccionario de mapeo de caracteres para reemplazar letras con tildes por letras sin tildes
mapeo_caracteres = {
    'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u',
    'Á': 'A', 'É': 'E', 'Í': 'I', 'Ó': 'O', 'Ú': 'U'
}

# Definir una función para reemplazar letras con tildes por letras sin tildes en un texto
def eliminar_tildes(texto):
    # Utilizar una expresión regular para encontrar letras con tildes y reemplazarlas utilizando el diccionario de mapeo de caracteres
    texto_sin_tildes = re.sub(r'[áéíóúÁÉÍÓÚ]', lambda x: mapeo_caracteres[x.group()], texto)
    return texto_sin_tildes

# Aplicar la función eliminar_tildes a cada texto en la lista datos_extraidos_total
datos_extraidos_sin_tildes = [eliminar_tildes(texto) for texto in datos_extraidos_total]

# Escribir los datos sin tildes en un nuevo archivo CSV
with open('datos_extraidos_sin_tildes.csv', 'w', newline='', encoding='utf-8-sig') as archivo:
    escritor_csv = csv.writer(archivo)
    for texto in datos_extraidos_sin_tildes:
        escritor_csv.writerow([texto])
```

```
import re
import csv

# Leer el archivo CSV con los datos extraídos sin tildes
with open('datos_extraidos_sin_tildes.csv', 'r', encoding='utf-8-sig') as archivo:
    lector_csv = csv.reader(archivo)
    # Extraer los datos de la primera columna en una lista
    datos_extraidos_sin_tildes = [fila[0] for fila in lector_csv]

# Definir la función para eliminar caracteres no deseados
def eliminar_caracteres(texto):
    # Definir una expresión regular para encontrar caracteres no alfabéticos o espacios
    patron = r'[^a-zA-ZáéíóúüñÁÉÍÓÚÜÑ\s]'
    # Reemplazar los caracteres no deseados por espacios en blanco
    texto_limpio = re.sub(patron, ' ', str(texto))
    return texto_limpio

# Aplicar la función eliminar_caracteres a cada texto en la lista datos_extraidos_sin_tildes
contenido_limpiado = [eliminar_caracteres(texto) for texto in datos_extraidos_sin_tildes]

# Convertir el texto limpiado a minúsculas
contenido_limpiado_minusculas = [texto.lower() for texto in contenido_limpiado]

# Escribir los datos limpiados y en minúsculas en un nuevo archivo CSV
with open('contenido_limpiado_minusculas.csv', 'w', newline='', encoding='utf-8-sig') as archivo:
    escritor_csv = csv.writer(archivo)
    for texto in contenido_limpiado_minusculas:
        escritor_csv.writerow([texto])
```

```
import nltk
from nltk.corpus import stopwords
import csv

# Leer el archivo CSV con los datos extraídos sin tildes
with open('contenido_limpiado_minusculas.csv', 'r', encoding='utf-8-sig') as archivo:
    lector_csv = csv.reader(archivo)
    # Extraer los datos de la primera columna en una lista
    contenido_limpiado_minusculas = [fila[0] for fila in lector_csv]

# Descargar la lista de stopwords si no la tienes
nltk.download('stopwords')

# Obtener la lista de stopwords en español
stopwords_espanol = stopwords.words('spanish')

# Definir una función para eliminar stopwords
def eliminar_stopwords(texto):
    # Tokenizar el texto en palabras
    palabras = texto.split()
    # Filtrar las palabras que no son stopwords
    palabras_filtradas = [palabra for palabra in palabras if palabra.lower() not in stopwords_espanol]
    # Unir las palabras filtradas en un texto nuevamente
    texto_filtrado = ' '.join(palabras_filtradas)
    return texto_filtrado

# Aplicar la función eliminar_stopword a cada texto en la lista
contenido_sin_stopword = [eliminar_stopwords(texto) for texto in contenido_limpiado_minusculas]

# Escribir los datos en un nuevo archivo CSV
with open('contenido_sin_stopword.csv', 'w', newline='', encoding='utf-8-sig') as archivo:
    escritor_csv = csv.writer(archivo)
    for texto in contenido_sin_stopword:
        escritor_csv.writerow([texto])

```

```
import nltk
from nltk.tokenize import word_tokenize
import csv

# Leer el archivo CSV con los datos extraídos sin stopwords
with open('contenido_sin_stopword.csv', 'r', encoding='utf-8-sig') as archivo:
    lector_csv = csv.reader(archivo)
    # Extraer los datos de la primera columna en una lista
    contenido_sin_stopword = [fila[0] for fila in lector_csv]

# Descargar los recursos necesarios si no los tienes
nltk.download('punkt')

# Definir una función para tokenizar el texto
def tokenizar_texto(texto):
    # Utilizar NLTK para tokenizar el texto en palabras
    tokens = word_tokenize(texto)
    return tokens

# Aplicar la función de tokenización a cada texto en la lista
contenido_tokens = [tokenizar_texto(texto) for texto in contenido_sin_stopword]

# Escribir los datos en un nuevo archivo CSV
with open('contenido_tokens.csv', 'w', newline='', encoding='utf-8-sig') as archivo:
    escritor_csv = csv.writer(archivo)
    for tokens in contenido_tokens:
        escritor_csv.writerow(tokens)
```

```
!pip install -U spacy
!python -m spacy download es_core_news_sm

import spacy
import csv

# Leer el archivo CSV con los datos tokenizados
with open('contenido_tokens.csv', 'r', encoding='utf-8-sig') as archivo:
    lector_csv = csv.reader(archivo)
    # Extraer los datos de cada fila en una lista de listas
    contenido_tokens = [fila for fila in lector_csv]

# Cargar el modelo de lenguaje en español
nlp = spacy.load('es_core_news_sm')

# Definir una función para lematizar el texto
def lematizar_texto(tokens):
    # Unir los tokens en una cadena de texto
    texto = ' '.join(tokens)
    # Procesar el texto con el modelo de lenguaje en español
    doc = nlp(texto)
    # Extraer las formas base (lemas) de cada palabra
    lemas = [token.lemma_ for token in doc]
    return lemas

# Aplicar la función de lematización a cada lista de tokens en la lista de contenidos
contenido_lemas = [lematizar_texto(tokens) for tokens in contenido_tokens]

# Escribir los datos en un nuevo archivo CSV
with open('contenido_lemas.csv', 'w', newline='', encoding='utf-8-sig') as archivo:
    escritor_csv = csv.writer(archivo)
    for lemas in contenido_lemas:
        escritor_csv.writerow(lemas)

```
